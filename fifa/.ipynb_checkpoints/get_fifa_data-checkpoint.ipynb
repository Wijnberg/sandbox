```python
# Import ma libraries
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as Soup
import time
```

```python
# Setup for scraper
releases = {'200020':'FIFA 2020', 
            '190075':'FIFA 2019',
            '180084':'FIFA 2018',
            '170099':'FIFA 2017',
            '160058':'FIFA 2016',
            '150059':'FIFA 2015',
            '140052':'FIFA 2014',
            '130034':'FIFA 2013',
            '120002':'FIFA 2012',
            '110002':'FIFA 2011',
            '100002':'FIFA 2010'}

columns = ['release','id','name','country','age','overall','potential','team','height',
           'weight','preferred_foot','best_position','value','wage']
df = pd.DataFrame(columns=columns)
```

```python
for release in list(releases.keys())[0:2]:
    for offset in range(0,1): #Should do 17

        # Fetch and parse
        url = f"https://sofifa.com/players?r={str(release)}&showCol%5B%5D=pi&showCol%5B%5D=ae&showCol%5B%5D=hi&showCol%5B%5D=wi&showCol%5B%5D=pf&showCol%5B%5D=oa&showCol%5B%5D=pt&showCol%5B%5D=bp&showCol%5B%5D=vl&showCol%5B%5D=wg&offset={str(offset*61)}"
        html = requests.get(url)
        soup = html.text
        data = Soup(soup,'html.parser')
        table = data.find('tbody')

        # Get all data from the table
        for i in table.find_all('tr'):    
            td = i.find_all('td')
            name = td[1].find_all('a')[1].text
            age = td[2].text.split()
            overall = td[3].find('span').text
            potential = td[4].find('span').text
            team = td[5].find('a').text
            id = td[6].text.strip()
            height = td[7].text.strip()
            weight = td[8].text.strip()
            preferred_foot = td[9].text.strip()
            best_position = td[10].text.strip()
            value = td[11].text.strip()
            wage = td[12].text.strip()

            # Get country of player by making new request
            '''
            player_url = f"https://sofifa.com/player/{id}"
            player_html = requests.get(player_url)
            player_soup = player_html.text
            player_data = Soup(player_soup,'html.parser')
            player_info = player_data.find_all("div", class_="info")
            country = player_info[0].find('a')['title']

            # Be polite
            time.sleep(1)
            '''

            # Put in a DataFrame and append
            data = [releases[release], id, name, country, age[0], overall, potential, team, height, 
                    weight, preferred_foot, best_position, value, wage]

            df_temp = pd.DataFrame([data])
            df_temp.columns = columns
            df = df.append(df_temp, ignore_index = True)
        
```

```python
for release in list(releases.keys())[0:2]:
    print(release)
```

```python
df.to_csv('fifa.csv', index=False)
```
